```python
import os,sys
sys.path.append(os.path.dirname(os.getcwd()))
from maesters import Maester
from datetime import datetime
```

```python
enfo = Maester('ecmwf','enfo',datetime(2022,6,25,12),hour=[6,30],varname='TP_L0')
icon = Maester('dwd','icon',datetime(2022,6,25,12),hour=[6,30],varname='TP_L0')
geps = Maester('cmc','geps_ens',datetime(2022,6,25,12),hour=[6,30],varname='TP_L0')
```

```python
enfo_data = enfo.xarray()
geps_data = geps.xarray()
icon_data = icon.xarray()
```

```python
enfo_data['TP_L0'].diff('time').sel(lon=slice(70,140),lat=slice(60,15)).plot(vmin=0,vmax=0.2)
```

```python
(icon_data['TP_L0'].diff('time').sel(lon=slice(70,140),lat=slice(15,60))/1000).plot(vmin=0,vmax=0.2)
```

```python
(geps_data['TP_L0'].diff('time').sel(lon=slice(70,140),lat=slice(15,60))/1000).plot(vmin=0,vmax=0.2)
```
